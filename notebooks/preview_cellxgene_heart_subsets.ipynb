# CellxGene Heart-Development Subset Preview

This notebook loads the 9 retained CellxGene heart-development subset `.h5ad` files and shows a small preview for each dataset.

For each subset, it displays:
- the dataset title and file name
- the dataset shape
- `adata.obs.head()`
- a tiny `5 x 5` cell-by-gene matrix preview


```python
from pathlib import Path
import warnings

import anndata as ad
import h5py
import pandas as pd
from IPython.display import Markdown, display

warnings.filterwarnings("ignore", category=FutureWarning)

SUBSET_DIR = Path("/gladstone/theodoris/lab/enockniyonkuru/maxtoki_development_data/source_1_cellxgene_heart_development_subset")

subset_files = sorted(
    path
    for path in SUBSET_DIR.glob("*_heart_development_subset.h5ad")
    if not path.name.startswith("02_")
)

assert len(subset_files) == 9, f"Expected 9 subset files, found {len(subset_files)}"

pd.DataFrame(
    {
        "subset_file": [path.name for path in subset_files],
        "path": [str(path) for path in subset_files],
    }
)
```

```python
def read_title(path: Path) -> str:
    with h5py.File(path, "r") as handle:
        uns = handle["uns"]
        if "title" in uns:
            value = uns["title"][()]
            if isinstance(value, bytes):
                value = value.decode("utf-8")
            elif hasattr(value, "decode"):
                value = value.decode("utf-8")
            title = str(value).strip().rstrip(".")
            return " ".join(title.replace("_", " ").split())
    fallback = path.name.replace("_heart_development_subset.h5ad", "")
    return " ".join(fallback.replace("_", " ").split())


def small_matrix_preview(adata, n_cells: int = 5, n_genes: int = 5) -> pd.DataFrame:
    preview = adata[:n_cells, :n_genes].X
    if hasattr(preview, "toarray"):
        preview = preview.toarray()
    return pd.DataFrame(
        preview,
        index=adata.obs_names[:n_cells],
        columns=adata.var_names[:n_genes],
    )


def preview_subset(path: Path, n: int = 5) -> None:
    title = read_title(path)
    adata = ad.read_h5ad(path, backed="r")
    try:
        display(Markdown(f"## {title}"))
        display(
            Markdown(
                f"- File: `{path.name}`  \n"
                f"- Shape: `{adata.n_obs:,} cells x {adata.n_vars:,} genes`"
            )
        )

        display(Markdown("**Observation metadata head**"))
        display(adata.obs.head(n))

        display(Markdown("**First 5 cells x first 5 genes**"))
        display(small_matrix_preview(adata, n_cells=n, n_genes=5))
    finally:
        adata.file.close()
```
